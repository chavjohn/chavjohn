Bar Chart visualization

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = df['Date']  # .columns('Date')
deaths_white = df['Deaths_White']    # .columns('Deaths_White')
deaths_black = df['Deaths_Black']    # .columns('Deaths_Black')


x = np.arange(len(labels))  # the label locations
width = 0.35 # the width of the bars

fig, ax = plt.subplots(figsize=(25,10))
rects1 = ax.bar(x - width/2, deaths_white, width, label='White')
rects2 = ax.bar(x + width/2, deaths_black, width, label='Black')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Deaths')
ax.set_xlabel ('Date')
ax.set_title('NY Covid19 Whites vs. Blacks Deaths from Jan to March 2021')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation= 45)
ax.legend(fontsize= 20)

def autolabel(rects):
  """Attach a text label above each bar in "rects", displaying its heigh. """
  for rect in rects:
      height = rect.get_height()
      ax.annotate('{}'.format(height),
                  xy=(rect.get_x() + rect.get_width() / 2, height), xytext=(0,1), # 3 points vertical offset 
                  textcoords="offset points",
                  ha='center',va='bottom')
      autolabel(rects1)
      autolabel(rects2)

      fig.tight_layout()

      plt.show()
```
